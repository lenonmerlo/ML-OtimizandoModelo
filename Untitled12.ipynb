# Conhecendo os dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/dados_inadimplencia.csv')
```

```python
dados.head()
```

```python
dados.info()
```

```python
dados['inadimplente'].value_counts(normalize=True)*100
```

```python
x = dados.drop('inadimplente', axis=1)
y = dados['inadimplente']
```

```python
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.33, random_state=RANDOM_STATE, stratify=y)
```

# Treinando o DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo_dt = DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE)
modelo_dt.fit(x_treino, y_treino)
```

```python
from sklearn.metrics import recall_score

recall_dt = recall_score(y_teste, modelo_dt.predict(x_teste))
print(f"Recall do DT = {recall_dt:.3f}")
```

# Treinando o LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
logistic_pipeline = make_pipeline(StandardScaler(), LogisticRegression())
```

```python
logistic_pipeline.fit(x_treino, y_treino)
```

```python
recall_lr = recall_score(y_teste, logistic_pipeline.predict(x_teste))
print(f"Recall do logistic regression = {recall_lr:.3f}")
```

# Busca em grade para o Decision Tree Classifier

```python
from sklearn.model_selection import GridSearchCV
```

```python
import numpy as np

param_grid_dt = {
    'criterion':  ['gini', 'entropy'],
    'max_depth': np.linspace(6, 12, 4, dtype=int),
    'min_samples_split': np.linspace(5, 20, 4, dtype=int),
    'min_samples_leaf': np.linspace(5, 20, 4, dtype=int),
    'max_features': ['sqrt', 'log2'],
    'splitter': ['best', 'random']
}
```

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(shuffle=True, random_state=RANDOM_STATE)
dt_grid_search = GridSearchCV(estimator=DecisionTreeClassifier(random_state=RANDOM_STATE),
             param_grid=param_grid_dt,
             scoring='recall',
             n_jobs=1,
             cv=cv)

dt_grid_search.fit(x_treino, y_treino)
```

# Como funciona a validação cruzada aninhada?

Conseguimos realizar a otimização de hiperparâmetros para o modelo DecisionTreeClassifier e para o LogisticRegression, testando diferentes combinações de hiperparâmetros, na busca de melhorar ainda mais o modelo inicial que construímos. Porém, no final da aula passada, ficou a dúvida: será que esse modelo está realmente nos dando a realidade ou será que eles estão otimistas demais?

A resposta é que, se utilizarmos a validação cruzada normal, a qual utilizamos na aula passada e que explicamos sucintamente, ela produz um resultado que é mais otimista do que deveria ser. Para resolver isso, utilizamos uma técnica para obter um erro de generalização mais adequado: a validação cruzada aninhada (Nested Cross Validation). Se abrirmos a documentação "Validação cruzada aninhada versus não aninhada", temos a seguinte frase:

"Nested CV estima o erro de generalização do modelo subjacente e sua pesquisa de hiperparâmetros. A escolha dos parâmetros que maximizam o CV não alinhado distorce o modelo em relação ao conjunto de dados, produzindo uma pontuação excessivamente otimista".

Nested CV é a validação cruzada aninhada. Então, resumidamente, se utilizarmos a validação cruzada não aninhada, o resultado é mais otimista do que realmente deveria ser. E, para estimar esse erro de generalização mais adequado, utilizamos o Nested CV (validação cruzada aninhada). A documentação explica como isso acontece:

"A seleção de um modelo sem CV alinhado usa os mesmos dados para ajustar os parâmetros do modelo e avaliar o desempenho do modelo. As informações podem, portanto, "vazar" para o modelo e ajustar demais os dados".

Portanto, um modelo sem CV alinhado, ou seja, anormal, é exatamente o que estávamos fazendo. Tínhamos uma parcela de dados para treino e outra para teste, e a partir disso, avaliávamos o nosso modelo. Como essas informações podem ser passadas para o modelo, podem produzir um resultado excessivamente otimista.

Entendemos, então, qual o problema de utilizar a validação cruzada não aninhada e quais as vantagens de utilizar a validação cruzada aninhada. Mas a pergunta que fica é: como que essa técnica funciona? Para responder isso, temos um slide explicando o passo a passo de como essa técnica é implementada.

A imagem mostra um diagrama explicativo do processo de validação cruzada aninhada, ou Nested Cross-Validation, para avaliação de modelo de machine learning. O diagrama é dividido em duas seções principais: a parte superior ilustra três iterações de validação cruzada externa, com divisões de conjuntos de treino e teste, e a parte inferior mostra a validação cruzada interna para escolha do melhor hiperparâmetro 'n_estimadores'. Cada iteração de validação cruzada interna tem duas divisões, uma para treino e outra para validação. Na terceira iteração, os dados de treino estão contornados de branco. Ao lado direito, os valores de 'recall' são apresentados para cada iteração, em letras brancas, e há um cálculo da média dos 'recalls' para as iterações externas e internas. As cores predominantes no diagrama são o verde-claro e verde-escuro em um fundo escuro com textos claros. Os recall que indicam as escolhas do melhor modelo estão marcados de verde.
Se tivermos uma validação cruzada aninhada, significa que teremos duas validações cruzadas, uma dentro da outra. Isso que significa o alinhado. Teremos o que chamamos de validação cruzada interna e a externa. A validação cruzada funciona da mesma forma como funcionou na aula anterior, a diferença é que vai ter uma dentro da outra.

Por exemplo, na validação cruzada nomeada de "1ª iteração" do nosso Nested Cross-Validation, apresentado no slide, separamos uma parcela de dados para treino e uma para teste. O mesmo acontece na segunda e terceira iteração. Por fim, fizemos a validação cruzada conforme conhecemos.

A diferença é que agora essa parcela de dados de treino, e lembrando que esse conjunto de dados já foi dividido, é o nosso conjunto de dados de treino. Não estamos mexendo no conjunto de dados de teste do train-test-split(). Essa parcela de dados é apenas o conjunto de dados de treino.

Na terceira iteração, pegamos somente essa parcela de dados que está sendo utilizada para treino dentro da validação cruzada. Essa parcela de dados que está sendo utilizada para treino, dentro da validação cruzada, será utilizada para outra validação cruzada, que é a validação cruzada interna. Então, pegamos essa parcela de dados e fazemos outra validação cruzada dentro dela.

Essa segunda validação cruzada, que está mais abaixo, será de fato utilizada para otimização de hiperparâmetros. Nela que determinamos qual o melhor hiperparâmetro, de acordo com essa média. Por exemplo, na segunda fase, onde temos n_estimadoress = 2, temos uma pasta de dados para Treino e outra para Validação, que gerou um RECALL 0.92. Outra pasta de validação e treino com esse mesmo n_estimadoress = 2 gerou uma RECALL 0.94.

Importante: No exemplo há dois quadrados para treino e dois quadrados para o CV interno, mas o CV interno e o CV externo podem ser de qualquer tamanho. Fica a seu critério.

Depois calculamos a métrica desse treino e validação e, para o hiperparâmetro n_estimadoress = 2 a média é 0.93. Em seguida, passamos para o próximo hiperparâmetro n_estimado = 5. Fazemos novamente treino e validação naquele conjunto de dados separado para treino, que é da terceira iteração e está separado com um contorno branco mais grosso. Ele devolve um RECALL 0.95 nos dois casos, então a média é 0.95. Portanto, o melhor é o com a média 0.95.

Dentro desse CV interno, escolhemos o melhor hiperparâmetro, no caso, n_estimadores = 5. Agora voltamos para o nosso CV externo, que agora avalia o erro de generalização do meu modelo. Temos o nosso modelo com n_estimadores = 5 e vamos treinar a parcela de dados de treino. Voltamos para a terceira iteração para testar nessa parcela de dados de teste, resultando em um RECALL 0.96.

Repetiremos esse processo para nas três iterações, apesar de só mencionar a terceira. A primeira teve um RECALL 0.98 e a segunda um RECALL 0.92. A partir disso calculamos a média desse nosso CV externo, que foi de 0.95. Então, para o meu modelo com n_estimadores = 5, o erro de generalização dele é de 0.95.

Isso é importante porque, na escolha de hiperparâmetros, nosso CV interno, com n_estimadores = 2 e n_estimadores = 5, ele não tem contato com essa parcela dos dados para teste no nosso CV externo. Então essas informações não vazam. Por isso que essa técnica é tão eficaz para estimar esse erro de generalização.

# Validação cruzada aninhada no Decision Tree Classifier

```python
inner_cv = StratifiedKFold(shuffle=True, random_state=RANDOM_STATE)

outer_cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_STATE)
```

```python
from sklearn.model_selection import cross_val_score

dt_nested_gs = GridSearchCV(estimator = DecisionTreeClassifier(random_state=RANDOM_STATE),
                                         param_grid = param_grid_dt,
                                         scoring = "recall",
                                         n_jobs = -1,
                                         cv = inner_cv
                                         )

dt_nested_scores = cross_val_score(dt_nested_gs, x_treino, y_treino, cv=outer_cv)
```

```python
print(f'resultado de cada iteração do cv externo: {dt_nested_scores}')
print(f'média: {dt_nested_scores.mean()}')
```

# Busca aleatória para o Decision Tree Classifier

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rs_dt = RandomizedSearchCV(estimator=DecisionTreeClassifier(random_state=RANDOM_STATE),
            param_distributions=param_grid_dt,
            n_iter=100,
            scoring='recall',
            n_jobs=-1,
            cv=inner_cv,
            random_state=RANDOM_STATE)

rs_dt_nested_score = cross_val_score(rs_dt, x_treino, y_treino, cv=outer_cv)
```

```python
rs_dt.fit(x_treino, y_treino)
rs_dt.best_params_
```

# Busca aleatória para o LogisticRegression

```python
rs_lr = RandomizedSearchCV(estimator=make_pipeline(StandardScaler(), LogisticRegression()),
        param_distributions=param_grid_lr,
        n_iter=50,
        scoring='recall',
        n_jobs=-1,
        cv=inner_cv,
        random_state=RANDOM_STATE)

rs_lr_nested_scores = cross_val_score(rs_lr, x_treino, y_treino, cv=outer_cv)

print(f'resultado de cada iteração do cv externo: {rs_lr_nested_scores}')
print(f'média: {rs_lr_nested_scores.mean()}')
```

```python
from skopt.space import Real, Integer, Categorical

space_dt = {
    'criterion': Categorical(['gini', 'entropy']),
    'max_depth': Integer(6, 12),
    'min_samples_split': Integer(5, 20),
    'min_samples_leaf': Integer(5, 20),
    'max_features': Categorical(['sqrt', 'log2']),
    'splitter': Categorical(['best', 'random'])
}
```

```python
from skopt import BayesSearchCV
```

```python
opt_dt = BayesSearchCV(estimator=DecisionTreeClassifier(random_state=RANDOM_STATE),
                    search_spaces = space_dt,
                    n_iter=50,
                    scoring='recall',
                    n_jobs=-11,
                    cv=inner_cv,
                    random_state=RANDOM_STATE)

bs_dt_nested_scores = cross_val_score(opt_dt, x_treino, y_treino, cv=outer_cv)
```

```python
print(f'resultado de cada iteração do cv externo: {bs_dt_nested_scores}')
print(f'média: {bs_dt_nested_scores.mean()}')
```

```python
opt_dt.fit(x_treino, y_treino)
opt_dt.best_params_
```
